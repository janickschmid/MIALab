```python
df = pd.read_csv('mia-result/plot/no-norm/results.csv', sep=';') #edit path accordingly
df['METHOD'] = pd.Series(['NONE' for x in range(len(df.index))])

comb = df
```

```python
 #add for each method
methods = {}
methods['Z-SCORE'] = 'plot/z-score_results'
methods['MINMAX'] = 'plot/minmax'
methods['LOG-NORM'] = 'plot/log_normalization'
methods['HIST-EQ'] = 'plot/hist_equalization'

for k,v in methods.items():
    add = pd.read_csv('mia-result/'+v+'/results.csv', sep=';')
    add['METHOD'] = pd.Series([k for x in range(len(add.index))])
    comb = comb.append(add)
```

```python
#Filter
amyg = pd.DataFrame(comb[comb.LABEL.isin(['Amygdala'])])
grey = pd.DataFrame(comb[comb.LABEL.isin(['GreyMatter'])])
hipp = pd.DataFrame(comb[comb.LABEL.isin(['Hippocampus'])])
thal = pd.DataFrame(comb[comb.LABEL.isin(['Thalamus'])])
whit = pd.DataFrame(comb[comb.LABEL.isin(['WhiteMatter'])])
```

```python
#Label comparison
for m in set(comb['LABEL']): 
     ax = comb[comb['LABEL']==m].boxplot(column='DICE',by=['METHOD'], figsize=(5,5))
     ax.set_title(m)
     plt.suptitle("")
```

```python
ax = comb.boxplot(column='DICE',by=['METHOD'], figsize=(5,5))
plt.suptitle("Overall performance")
```
